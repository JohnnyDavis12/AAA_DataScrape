```python
df = pd.read_csv("/Users/johndavis/Desktop/AAA_data.csv",low_memory=False)
```

```python
#unique Game date
df['game_date'].unique()
```

```python
dft = df[['game_pk','game_date','matchup.pitcher.fullName','matchup.pitchHand.code', 'home_parentOrg_name','away_parentOrg_name','about.isTopInning','details.description','result.event','result.eventType','details.type.code','pitchData.startSpeed','pitchData.breaks.spinRate','pitchData.breaks.breakVerticalInduced','pitchData.breaks.breakHorizontal','pitchData.coordinates.pZ','pitchData.extension','pitchData.coordinates.pX','pitchData.coordinates.pZ']]
dft.columns = ['game_pk','game_date','Pitcher','pitcher_hand','home_team','away_team','top_inning','pitch_description','event','event_type','TaggedPitchType','RelSpeed','SpinRate','InducedVertBreak','HorzBreak','pZ','Extension','PlateLocSide','PlateLocHeight']
```

```python
#when inning_topbot = Bot	PitcherTeam = away_team and when inning_topbot = Top PitcherTeam = home_team
dft['PitcherTeam'] = np.where(dft['top_inning'] == True, dft['home_team'], dft['away_team'])
```

```python
dft['TaggedPitchType'] = dft['TaggedPitchType'].replace(['FF','SI','FS','FA'],'Fastball')
dft['TaggedPitchType'] = dft['TaggedPitchType'].replace(['SL','ST','FC'],'Slider')
dft['TaggedPitchType'] = dft['TaggedPitchType'].replace(['CH','EP','SC','FO','KN'],'ChangeUp')
dft['TaggedPitchType'] = dft['TaggedPitchType'].replace(['SV','CU','KC'],'Curveball')
```

```python
dft['ABS_Horzizontal'] = abs(dft['HorzBreak'])
dft['differential_break'] = abs(dft['InducedVertBreak'] - dft['ABS_Horzizontal'])
```
